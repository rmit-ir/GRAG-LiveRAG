```python
import pandas as pd

df_ref = pd.read_csv(
    '../data/generated_qa_pairs/dmds_1GMhaG2S7ONAd.multi.n50.tsv', sep='\t', dtype={'qid': int}).rename(
        columns={'question': 'ref_question', 'answer': 'ref_answer', 'document_ids': 'ref_doc_ids', 'context': 'ref_context'})
df_rag_results = pd.read_csv(
    '../data/rag_results/dmds_1GMhaG2S7ONAd_BasicRAG2.tsv', sep='\t', dtype={'qid': int})
df_eval = pd.read_csv(
    '../data/evaluation_results/dmds_1GMhaG2S7ONAd_BasicRAG2.LLMEvaluator.eval.rows.tsv', sep='\t',)
```

```python
df_rag_results.head(5)
```

```python
df_results = pd.concat([df_ref.set_index('qid'),
                        df_rag_results.set_index('qid'),
                        df_eval.set_index('qid')], axis=1)
df_results_filtered = df_results[['question', 'answer', 'ref_answer', 'relevance_score', 'faithfulness_score', 'generated_queries', 'context', 'ref_context']]
df_results_filtered.to_csv('../data/evaluation_results/manual_inspection_results.tsv', sep='\t', index=True)
df_results_filtered.to_excel('../data/evaluation_results/manual_inspection_results.xlsx')
```
